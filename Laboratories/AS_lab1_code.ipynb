```python
#4
#Hypothesis
#H0: Variances in both group of salaries are the same.
# Fisher test
F = (data['over.1year'].var())/(data['not.1year'].var())
df1 = len(data['over.1year'])-1
df2 = len(data['not.1year'])-1
alpha = 0.05
p_value = stats.f.cdf(F, df1, df2)

if p_value > alpha:
    print(p_value,'>',alpha)
    print(True)
elseą
    print(p_value,'<',alpha)
    print(False)
```

Commentary: Becouse p_value > alpha so the data do not provide sufficient evidence to reject the null hypothesis.

```python
#5
#Hypothesis
#Student test
#H0: Average payment in the group of employees 'over.1year' is the same as in the group of 'not.1year'.

results = stats.ttest_ind(data['over.1year'],
                data['not.1year'])
print(results)
```

Commentary:
The p value is lower than p < 0,05, so we can reject the null hypothesis.
